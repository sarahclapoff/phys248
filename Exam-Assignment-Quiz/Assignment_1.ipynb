## Problem 3
The one-dimensional equation of motion of a ball with an initial velocity $v_\mathrm{0}$ in the positive direction (up) is given by
$$
y_\mathrm{b} = -\frac{1}{2}gt^2 + v_\mathrm{0} t 
$$
This equation gives the height of the ball for a given time, which represents the time passed since the ball has been thrown in the upward direction.

1. Write a code cell that evaluates the mathematical expression for $y_\mathrm{b}$ for $t_1 = 2.7\mathrm{s}$ and prints out the result in a formatted print statement `yb = x.yy`. 
2. Evaluate the expression for 21 equally spaced times from $0.$ to $3t_1$ ($t \in [0.,2t_1]$, unit of time is s). Save the t and $y_\mathrm{b}$ values in the arrays `times` and `yb`. Avoid using a loop.
3. Make a line plot of $y_\mathrm{b}(t)$. By looking at your graph determine if the ball is on the way up or down at $t_1$? 
4. Confirm this finding by calculating the gradient $d y_\mathrm{b}/dt$ using the numpy `gradient` function and determining the sign of the gradient for $t=t_1$. Write you finding in a print statement. 
4. What is the maximum height the ball reaches? 


**Variables:**
```Python
g = 9.81    # m s^-2
v0  = 19.1  # m s^-1
```

````python
#part 1
g = 9.81 #m/s^2
v0 = 19.1 #m/s
t1 = 2.7 #s
y1 = -1/2*g*t1**2 + v0*t1 #m/s
print('1. yb(%.1f) = %.2f' %(t1, y1))
print()

#part 2
n = 21
times = np.linspace(0, 3*t1, n)
yb = -1/2*g*times**2 + v0*times #m/s

#part 3
plt.plot(times, yb)
plt.xlabel("Time")
plt.ylabel("Height")
plt.title("Ball Height vs. Time")
plt.show()
print("2. From the plot above, the ball is going down at t1 = %.1fs." %(t1))
print()

#part 4
grad = np.gradient(yb)
t_a = int(n/3)-1
t_b = int(n/3)
#the times array doesn't have t1 as an element, so consider the times closest to t1.
#range of times is (0, 3*t1), so the times element closest to t1 is n/3.
print("3. The gradient at t=%.1fs is %.3f, and the gradient at t=%.1fs is %.3f." %(times[t_a], grad[t_a], times[t_b], grad[t_b]))
print("Considering the above plot, the trajectory of the ball doesn't change between these times, so the sign of the gradient for t=%.1fs is negative." %(t1))
print()

#part 5
print('4. Max hight the ball reaches is %.4fm' %(max(yb)))
````

### 4.1
Write a program that adds all integer number $i \in [0,10000]$ excluding those that can be divided by $7$ or $5$. Solve this problem with a `for` loop and then without a loop, but instead using vector operations.

````python
#with for-loop
sum1 = 0
for i in range(10000):
    if (i%7 != 0) and (i%5 != 0):
        sum1 += i
print("Sum with for-loop:", sum1)

#with vector operations
nums = np.arange(0,10001, 1)
nums = np.delete(nums, np.where(nums%7 == 0))
nums = np.delete(nums, np.where(nums%5 == 0))
sum2 = sum(nums)
print("Sum with vector operations:", sum2)
````
