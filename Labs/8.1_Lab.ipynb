Complete  Activity 1 before the end of the lab by entering the answer into the Brigthspace quiz.

Solve the following differential equation numerically using `scipy.odeint` in the intervall $x \in [0.5,8]$ and $y(0.5) = 2$.
$$
\frac{dy}{dx} = 2 x \sin(x) \, e^\frac{-2x}{3}
$$
and plot $y(x)$. Use an  array for $x$ with $31$ equidistant points.


Multiply the maximum of $y$ and the $x$ value at which $y$ has the maximum, round to two decimals and enter into quiz `max-ode1`.

**Solution:**

```python
from scipy import integrate
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dydx(y,x):
    return 2*x*np.sin(x)*np.exp(-2*x/3)

y0 = 2
x = np.linspace(0.5, 8, 31)
y = integrate.odeint(dydx, y0, x)

#ravel() gets all ur individual elements into one array
#print(y)
#print(y.ravel())
y=y.ravel()

plt.plot(x,y)

#index of ymax:
ymax_index = np.argmax(y)
xmax = x[ymax_index]

max_ode1 = max(y) * xmax
print("max_ode1: %.2f"%max_ode1 )
```

### Activity 2
Name: `poly6-rootsum`

$$
P(x) = 924x^6 -2772x^5+3150x^4 -1680 x^3 + 420 x^2 -42 x +1
$$

1. Make a plot of $P(x)$ for $ x\in[0,1]$. By inspecting it find approximate values for the six roots. 
2. Write a program to solve for the positions of all six roots, using the Newton-Raphson method (you can use a library if you want). Add the roots and enters the resulting number into question 

**Solution:**

```python
def P(x):
    return 924*x**6 - 2772*x**5 + 3150*x**4 - 1680*x**3 + 420*x**2 - 42*x +1

x = np.linspace(0,1,101)
P_arr = P(x)
plt.plot(x, P_arr)
plt.grid()
plt.title("$P(x)$")
plt.ylabel("$P(x)$")
plt.xlabel("$x$")
plt.show()

root_init = [.03, .16, .38, .62, .83, .97]

y_roots = np.zeros(6)
x_roots = optimize.newton(P, root_init)
plt.plot(x, P_arr, label = 'P(x)')
plt.scatter(x_roots, y_roots, label='roots of P(x)')
plt.grid()
plt.legend()
plt.show()

print("The roots of P(X) are ", x_roots)
print("poly6-rootsum = ", sum(x_roots))
```
