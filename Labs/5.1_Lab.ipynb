Name: `int-midpoint`

1. Create a python function using the `def` method that takes `x` as an input and returns the function value 
$$f(x) =  x^2 e^x
$$ 
for $x\in[-1,1]$. 
2. Using this function create a bar plot with 20 rectangles according to the midpoint rule. Also add a line plot of $f(x)$ with an array for plotting `xp` that has 200 points.
3. Calculate the area under the curve by numerical integration using the midpoint rule. Don't use a loop. Print the result in a formatted print statement, with 4 decimals. Enter just the numerical answer into the Brighspace question `int-midpoint`.



**Solution:**

```python
%pylab ipympl
```

```python
def func(x):
    return x**2 * e**x

#2
x= linspace(-1,1,20)
delta_x = diff(x)[0]
xp = linspace(-1,1,200)
ifig=1; close(ifig); figure(ifig)
bar(x, func(x), delta_x)
plot(xp, func(xp), '--', color='pink')

#3
func_area = sum(func(x)*delta_x)
print("Approximate int(func(x)) for x in [-1,1] = {:6.4f}".format(func_area))
```

```python
%%bash

head -1 owid-covid-data.csv > canada-covid.csv
grep "Canada" owid-covid-data.csv >> canada-covid.csv
```

```python
%pylab ipympl
import csv
from datetime import datetime

dates = []
new_cases = []
file = open("canada-covid.csv","r")
data = csv.reader(file)

for i, row in enumerate(data):
    if i != 0:
        dates.append(row[3])
        new_cases.append(row[5])
new_cases = np.array(new_cases, dtype = float)
file.close()

days_offset = [datetime.strptime(date,"%Y-%m-%d").timestamp() for date in dates]
days_offset = array(daysoffset) - daysoffset[0]
days_offset = daysoffset/(24*3600)

figure(figsize=(12,5))
plot(days_offset,new_cases)
title("new cases vs. days offset")
```

```python
smooth_newcases = [] 
for i in range(7, len(new_cases)-6):
    smooth_newcases.append(np.mean(new_cases[i-7:i+7]))

smooth_days = days_offset[7:-6]
    
figure(figsize=(12,5))
plot(smooth_days, smooth_newcases)
title("smoothed new cases vs. days")
print('max num new cases = ', max(smooth_newcases))
```
